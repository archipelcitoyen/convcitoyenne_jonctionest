```python
import random
import math
import folium
```

```python
# Nombre de PàP prévu
num_points = 20

# Localisation du centre (latitude, longitude in decimal degrees)
center_lat = 43.587  # 43°35'13.2"N
center_lon = 1.4957  # 1°29'44.5"E

# Rayon du cercle
rayon = 3
```

```python
# Earth's radius in kilometers
R = 6371

# Function to generate random points within a given radius (in kilometers)
def random_points_in_circle(center_lat, center_lon, radius_km, num_points):
    points = []

    # Convert the latitude and longitude offsets for the specific location
    lat_per_km = 1 / 111  # Approximate km per degree latitude
    lon_per_km = 1 / (111 * math.cos(math.radians(center_lat)))  # Adjust for the latitude

    for _ in range(num_points):
        # Generate a random distance within the circle
        d = radius_km * math.sqrt(random.uniform(0, 1))

        # Random angle (bearing) in radians
        theta = random.uniform(0, 2 * math.pi)
        print(f'')

        delta_lat = d * math.sin(theta) * lat_per_km
        delta_lon = d * math.cos(theta) * lon_per_km

        new_lat = center_lat + delta_lat
        new_lon = center_lon + delta_lon

        points.append((new_lat, new_lon))
        print(f'd={d} \ttheta={theta} \tlat={new_lat} \tlon={new_lon}')

    return points
```

```python
# Step 2: Generate the random points
random_points = random_points_in_circle(center_lat, center_lon, rayon, num_points)
```

```python
# Carte
mymap = folium.Map(location=[center_lat, center_lon], zoom_start=13)
folium.Marker([center_lat, center_lon], popup="Center Point").add_to(mymap)
for i, (lat, lon) in enumerate(random_points, 1):
    folium.Marker([lat, lon], popup=f"Point {i}").add_to(mymap)
mymap
```
